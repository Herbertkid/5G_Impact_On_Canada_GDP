```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dftotalGDP = pd.read_csv('/content/drive/MyDrive/Sys5804 project/project/totalGDP.csv')
```

```python
dftotalGDP['Year']
```

```python

fig, ax = plt.subplots()

ax.plot(dftotalGDP['Year'], dftotalGDP['CAGDP'], label='CA') 
ax.plot(dftotalGDP['Year'], dftotalGDP['ONGDP'], label='ON') 
ax.plot(dftotalGDP['Year'], dftotalGDP['QCGDP'], label='QC') 
ax.plot(dftotalGDP['Year'], dftotalGDP['ABGDP'], label='AB') 
ax.set_xlabel('Year') 
ax.set_ylabel('GDP') 
ax.set_title('GDP prediction') 
ax.legend() 

plt.show() 

```
